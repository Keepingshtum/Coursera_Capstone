## Acquiring the table from Wikipedia

We can get the table and its data with the help of a little pandas magic:

```python
table=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
df=table[0]
df.head()
```

We can see that some values are **not assigned**. Also, some addresses list **two different neighbourhoods**. We must address that!

```python
#find the index of all codes not assigned, and drop them.
df_drop1=df[df['Borough']=='Not assigned']
df.drop(df_drop1.index,axis=0,inplace=True) 
```

```python
#Assign Neighbourhood the same as Borough; where neighbourhood is not assigned, but the borough is:
df_up1=df[df['Neighbourhood']=='Not assigned']
df_up1.Neighbourhood=df_up1.Borough
```

```python
#update the values.
df.update(df_up1)
```

```python
#club the the neighbourhoods sharing the same postal code
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df.head()
```

# Visualising the data

Now that we have cleaned up the data, let us examine how the postal codes are actually present in a map of Toronto:

```python
#read geospatial coordinates from file and add them to the dataframe
df2=pd.read_csv('Geospatial_Coordinates.csv')
df['Latitiude']=df2['Latitude']
df['Longitude']=df2['Longitude']
```

```python
df["Latitiude"]
```

```python
import sys
import folium
```

```python
map_toronto = folium.Map(location=[43.653963, -79.387207], zoom_start=11)
for lat,lon,name in zip(df['Latitiude'],df['Longitude'],df['Borough']):
    folium.Marker(location=[lat, lon],popup=name,parse_html=True).add_to(map_toronto)
map_toronto.save("markers_on_folium_map1.html")
map_toronto
```
